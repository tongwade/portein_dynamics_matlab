```python
all = pd.read_csv('/home/tongwade780/pdb_website/python_code/all_cluster_feature_csv_add_feature/all_clusters_all_features_add_3_features_clusterID_pca_3.csv')
```

```python
import numpy as np
have_pdb = []
for cluster in range(0,1673):
    cluster_pdb_name = pdb_name[cluster]
    cluster_pdb_name = np.asarray(cluster_pdb_name)
    cluster_pdb_name = filter(None, cluster_pdb_name)
    cluster_num = cluster+1
    
    pca_distance_matrix = h5py.File('/mnt/Tsunami_HHD/Tang/onebyone_clsuter/cluster_pca_distance/cluster_{0}_pca_distance.mat'.format(cluster_num),'r')
    PCA_num = pca_distance_matrix['all_dis'].value.T

    dist_1 = PCA_num[0:,0]
    dist_2 = PCA_num[0:,1]
    dist_3 = PCA_num[0:,2]    
    
    test_dict = zip( map(float,dist_1), cluster_pdb_name)
    new_dict = []
    for line in test_dict:
        if np.isnan(line[0]):   
            1
        else:
            new_dict.append(line)
    new_dict.sort()
    for one_dict in new_dict:
        if len(all[all.PDB_ID.isin([one_dict[1][0:4].lower()])]) != 0:
            have_pdb.append((one_dict[1][0:4].lower(),cluster_num))
            break    
            
    test_dict = zip( map(float,dist_2), cluster_pdb_name)
    new_dict = []
    for line in test_dict:
        if np.isnan(line[0]):   
            1
        else:
            new_dict.append(line)    
    for one_dict in new_dict:
        if len(all[all.PDB_ID.isin([one_dict[1][0:4].lower()])]) != 0:
            have_pdb.append((one_dict[1][0:4].lower(),cluster_num))
            break    
            
    test_dict = zip( map(float,dist_3), cluster_pdb_name)
    new_dict = []
    for line in test_dict:
        if np.isnan(line[0]):   
            1
        else:
            new_dict.append(line)    
    for one_dict in new_dict:
        if len(all[all.PDB_ID.isin([one_dict[1][0:4].lower()])]) != 0:
            have_pdb.append((one_dict[1][0:4].lower(),cluster_num))
            break   

import pandas as pd
have_pdb_data=pd.DataFrame(have_pdb)
have_pdb_data.columns = ['pdbname','cluster']

have_pdb_data.to_csv('/home/tongwade780/pdb_website/python_code/all_cluster_problem_pdb_add_feature/pca_3_mode_pdb_name_new.csv',index= False)            
```
