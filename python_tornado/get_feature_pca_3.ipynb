```python
import pandas as pd

all = pd.read_csv('/mnt/Tsunami_HHD/Tang/all_feature_data/cluster_2_all_feature.csv')
with open('/home/tongwade780/pdb_website/python_code/all_cluster_feature_csv_new/onechain_100clust.csv','r') as f:
    lines = f.readlines()

clusters = [map(lambda x: x[0:4],line.strip().split(',')) for line in lines]

dd = {}
for i,cluster in enumerate(clusters):
    for pdbID in cluster:
        if pdbID:
            dd[pdbID] = i+1

all['clusterID'] = all['PDB_ID'].apply(lambda x: dd[x.upper()])            
```
