```python
with open('/home/tongwade780/pdb_website/python_code/all_cluster_feature_csv_new/onechain_100clust.csv','r') as f:
    lines = f.readlines()

clusters = [map(lambda x: x[0:4],line.strip().split(',')) for line in lines]

dd = {}
for i,cluster in enumerate(clusters):
    for pdbID in cluster:
        if pdbID:
            dd[pdbID] = i+1

all['clusterID'] = all['PDB_ID'].apply(lambda x: dd[x.upper()])      
```

```python
change_order = ['PDB_ID','resID','atomic_1','atomic_3','atomic_5','ssbond_matrix','Polarity',
                'eig_vv_1','eig_vv_2','vector_1','Rank_1','vector_2','Rank_2',
                'GNM_co','co_rank','charge','stand_Dcom','side_chain_length','pka_charge',
                'dssp_result','sasa','entropy','ca_len','hole_ss_bond','rmsf_overall',
                'contact_per_res','gyradius','pca_feature','eig_first','eig_sec',
                'shannon_entropy','shannon_entropy20','eig5_eig1','phobic_percent',
                'loop_percent','h_bond_group','intrinsicDisorder','pca_feature_2','pca_feature_3','sheet_percent','helix_percent','RMSF','RMSF_0f_bfactor',
                'shifted_RMSF','clusterID']
all = all[change_order]
```
